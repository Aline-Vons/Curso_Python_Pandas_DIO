```python
#criando o dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#Qual foi a receita total?
df["Valor Venda"].sum()
```

```python
#Qual o custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #criando a coluna de custo
```

```python
df.head()
```

```python
#qual o custo total?
round(df["custo"].sum(), 2) #usa o round para arredondar
```

```python
#lucro receita menos o custo
df["Lucro"] = df["Valor Venda"] - df["custo"] #criamos a coluna
```

```python
df.head(1)
```

```python
round(df["Lucro"].sum(), 2)
```

```python
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days #usamos para remover o days e deixar somente o número
```

```python
df.head(1)
```

```python
#verificando o tipo de uma coluna especifica
df["Tempo Envio"].dtype
```

```python
#encontrando a média do tempo de envio por marca
df.groupby("Marca")["Tempo Envio"].mean()
```

**MISSING VALUES**

```python
#verificando se falta dados - valores ausentes
df.isnull().sum()
```

```python
#lucro por ano e por marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum() #não precisou criar uma coluna ano por conta do .dt.year
```

```python
#para tirar a notação cientifica e configurar o float
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
```

```python
lucro_ano
```

```python
#Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receitas");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head ()
```

```python
#lucro por mes
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
#lucro por marca
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal'); #horizontal no caso são as legendas do gráfico
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
