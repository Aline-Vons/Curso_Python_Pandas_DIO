```python
#As tuplas usam parentesis como sintaxe
tp = ("banana", "maçã", 10, 50)
```

```python
#retornando o primeiro elemento
tp[0]
```

```python
#diferente das listas as tuplas são imutáveis
tp[0] = "laranja"
```

```python
tp.count("maçã")
```

```python
tp[0:2]
```

**DICIONÀRIOS**

```python
#para criar um dicionário usamos as {}. Trabalham com o conceito de chave e valor
dc = {"maçã":20, "banana":10, "laranja":15, "uva":5}
```

```python
dc
```

```python
#acessando o valor de um dicionário através da chave
dc["maçã"]
```

```python
#retornando todos as chaves do dicionário
dc.keys()
```

```python
#retornando todos os valores do dicionário
dc.values()
```

```python
#verificando se já existe uma chave e caso não exista, inserir
dc.setdefault("limão", 22)
```
