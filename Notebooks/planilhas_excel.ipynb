**Trabalhando com planilhas Excel**

```python
#passo 1: importar a biblioteca panda
import pandas as pd
```

```python
#leitura dos arquivos excel
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df1.head()
```

```python
#juntando todos os arquivos
df= pd.concat([df1,df2,df3,df4,df5])
```

```python
#exibir as 5 primeiras linhas
df.head()
```

```python
#exibir as últimas linhas
df.tail()
```

```python
#mostra do conjunto de dados
df.sample(5)
```

```python
#verificando o tipo de dado de cada coluna
df.dtypes
```

```python
df.head()
```

```python
#alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"]. astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
df.isnull()
```

```python
#substituindo os valores nulos por zero##inplace serve para substituir definitivo os valores alterados###
df["Vendas"].fillna(0, inplace= True)
```

```python
#apagando linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos cm base em apenas uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo as linhas com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

**Criando colunas novas**

```python
#multiplicando a coluna vendas x quantidade (mul) para gera nova de Receitas
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#se precisasse saber o valor de receitas/vendas
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#encontrando maior receita
df["Receita"].max()
```

```python
#encontrando menor receita
df["Receita"].min()
```

```python
#encontrando o top 3 de Receitas
df.nlargest(3, "Receita")
```

```python
#encontrando o top 3 de menor Receitas
df.nsmallest(3, "Receita")
```

```python
#agrupando por cidade x receita
df.groupby("Cidade")["Receita"].sum()
```

```python
##ordenando nossos dados
df.sort_values("Receita", ascending=False).head(10)
```

**Trabalhando com datas**

```python
#transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
#transformando a coluna de data em data (pandas (pd) para datetime)
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#extrair o mês e o dia e criando nova coluna
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#retornando data mais antiga
df["Data"].min()
```

```python
#calculando a diferença de dias (por não ter 2 colunas de datas vamos subtrair da data mais antiga)
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#criando coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#Filtrando por ano e mês específico
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
vendas_marco_19
```

```python
#adicionando titulo e alterando nome dos eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
df.groupby(df["mes_venda"])["Qtde"].sum().plot() #agrupado mes de venda e quantidade, aplicada a soma. o .plot é para criar grafico de linhas
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend() #pedir para aparecer legenda
```

```python
df.groupby(df["mes_venda"])["Qtde"].sum().plot(title = "Total Produtos Vendidos x Mês") #title para por legenda
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend() #pedir para aparecer legend
```

```python
#selecionando apenas vendas de 2019
df_2019 = df[df["Ano_Venda"] == 2019]
```

```python
#Total produtos vendidos por mês
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v") #marcadores *. o, etc..
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend(); #; é usado para não mostrar linha cabeçalho
```
